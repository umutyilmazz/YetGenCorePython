# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self,brand,model,year,isManuel):
        self.brand = brand
        self.model = model
        self.year = year
        self.isManuel = isManuel
    def printQualities(self):
        print(f"""
            Vehichle's Properties:
                Brand: {self.brand}
                Model: {self.model}
                Year: {self.year}
                Manuel: {self.isManuel}
        """)

class Bus(Vehicle):
    def __init__(self, brand, model, year, isManuel):
        super().__init__(brand, model, year, isManuel)

    def seating_capacity(self, capacity = 50):
        print(f"The seating capacity of a bus is {capacity} passengers")

mercedes = Bus("Mercedes","Travego",2020,True)
mercedes.printQualities()
mercedes.seating_capacity()
````

# Problem 2


School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
# 2.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self,brand,model,year,isManuel):
        self.brand = brand
        self.model = model
        self.year = year
        self.isManuel = isManuel
    def printQualities(self):
        print(f"""
            Vehichle's Properties:
                Brand: {self.brand}
                Model: {self.model}
                Year: {self.year}
                Manuel: {self.isManuel}
        """)

class Bus(Vehicle):
    def __init__(self, brand, model, year, isManuel):
        super().__init__(brand, model, year, isManuel)

    def seating_capacity(self, capacity = 50):
        print(f"The seating capacity of a bus is {capacity} passengers")

class School_bus(Bus):
    def __init__(self, brand, model, year, isManuel):
        super().__init__(brand, model, year, isManuel)
    
    def seating_capacity(self, capacity=30):
        return super().seating_capacity(capacity)

sprinter = School_bus("Mercedes","Sprinter",2019,False)
sprinter.printQualities()
sprinter.seating_capacity()
````
