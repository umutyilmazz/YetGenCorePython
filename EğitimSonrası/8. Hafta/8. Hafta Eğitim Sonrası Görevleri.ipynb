# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import time
def vkiHesabi(boy,kilo):
    return kilo / (boy/100) ** 2
def degerOlusturma(vki):
    if vki < 18.5:
        return "Zayıf"
    elif vki < 24.9:
        return "Normal"
    elif vki < 29.9:
        return "Fazla Kilolu"
    elif vki < 39.9:
        return "Obez"
    else:
        return "İleri Derecede Obez (Morbid Obez)"
def cikti(isim,yas,boy,kilo,vki, deger):
    tarih = time.ctime(time.time())
    deger1 = f"Kullanıcının Adı Soyadı: {isim}\n"
    deger2 = f"Kullanıcının Yaşı, Boyu(cm), Kilosu: {yas},{boy},{kilo}\n"
    deger3 = f"Kullanıcının VKİ Değeri: {vkiHesabi(boy,kilo)}\n"
    deger4 = f"Değer Aralığı: {degerOlusturma(vki)}\n"
    deger5 = f"Oluşturulma Tarihi: {tarih}\n"
    output = deger1 + deger2 + deger3 + deger4 + deger5
    return output
def main():
    sayi = int(input("Kaç Kullanıcı Gireceksiniz: "))
    for x in range(sayi):
        print(f"{x+1}. Kullanıcı Kaydı")
        isim = input("Kullanıcı Ad Soyadı: ")
        yas = int(input("Kullanıcı Yaşı: "))
        kilo = float(input("Kullancı Kilosu: "))
        boy = float(input("Kullanıcı Boyu(cm): "))
        vki = vkiHesabi(boy,kilo)
        deger = degerOlusturma(vki)
        output = cikti(isim,yas,boy,kilo,vki,deger)
        with open("vkiDegerleri.txt","a",encoding="utf-8") as f:
            f.write(output)
            f.write("-"*50)
            f.write("\n")
            print("Kullanıcı Başarıyla Kaydedildi!")
main()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import time
anaMalzeme = {
    "elma" : 0.5,
    "armut" : 0.75,
    "muz" : 2,
    "çilek" : 2,
    "erik" : 3,
    "şeftali" : 1,
    "salatalık" : 1,
    "domates" : 1,
    "ıspanak" : 1
}
yardimciMalzeme = {
    "pastorize süt" : 2,
    "laktozsuz süt" : 4,
    "badem sütü" : 6,
    "yağsız süt" : 4 
}
ekstraMalzeme = {
    "tarçın" : 1,
    "keten tohumu" : 4,
    "zencefil" : 2
}
def menuGor():
    print("YetBar Menü".center(50,"-"))
    print("Ana Malzemeler".center(50,"-"))
    for x in anaMalzeme:
        print(f"{x} : {anaMalzeme[x]} TL")
    print("Yardımcı Malzemeler".center(50,"-"))
    for y in yardimciMalzeme:
        print(f"{y} : {yardimciMalzeme[y]} TL")
    print("Ekstra Malzemeler".center(50,"-"))
    for z in ekstraMalzeme:
        print(f"{z} : {ekstraMalzeme[z]} TL")
def fiyatHesabi(urun,adet):
    assert adet >= 0
    if not urun in anaMalzeme or not urun in yardimciMalzeme or not urun in ekstraMalzeme:
        return None
    elif urun in anaMalzeme:
        return anaMalzeme[urun] * adet
    elif urun in yardimciMalzeme:
        return yardimciMalzeme[urun] * adet
    else:
        return ekstraMalzeme[urun] * adet
def satinAlim():
    menuGor()
    print("Satın Almak İstediğiniz Ürünü Giriniz(Çıkış İçin 0'a Basınız): ")
    urunler = []
    adetler = []
    toplam = 0
    while 1:
        menuSecim = int(input("Hangi Menüden Seçim Yapacaksınız (Ana Malzeme: 1 | Yardımcı Malzeme: 2 | Ekstra Malzeme: 3): "))
        if menuSecim == "0":
            print("Fiş Kaydediliyor...")
            break
        assert menuSecim in "123"
        if menuSecim == "1":
            urunSecim1 = input("Hangi Üründen Ekleyeceksiniz: ")
            urunAdet = input("Kaç Adet Almak İstiyorsunuz: ")
            fiyat = fiyatHesabi(urunSecim1,urunAdet)
            if fiyat is None:
                print("Yanlış Bir Değer Girdiniz")
            else:
                print(f"{fiyat} TL Fiyatı Onaylamak İçin 'y' iptal etmek için 'n' basınız")
                check = input("Satın Almayı Onaylıyor Musunuz? ")
                if check == "y":
                    print("Satın Alma Onaylandı...")
                    print("-"*50)
                    toplam += fiyat
                    urunler.append(urunSecim1)
                    adetler.append(urunAdet)
                    print(f"Ara Toplam: {toplam} TL")
                elif check == "n":
                    print("Satın Alma İşlemi İptal Edildi")
                    print("-"*50)
    return toplam, urunler, adetler
def main():
    print("0-Çıkış\n1-Menü Gör\n2-Satın Alım")
    islem = None
    while 1:
        islem = int(input("İşlem Tercihi: "))
        if islem < 0 or islem > 2:
            print("Hatalı İşlem")
        elif islem == "0":
            print("Çıkış Yapılıyor...")
            break
        elif islem == "1":
            menuGor()
        elif islem == "2":
            toplamFiyat , urunler, adetler = satinAlim()
            with open("yetBarVerileri.txt","a",encoding="utf-8") as f:
                f.write(str(time.ctime(time.time())))
                f.write("\n")
                f.write(",".join(map(str,urunler)))
                f.write("\n")
                f.write(",".join(map(str,adetler)))
                f.write("\n")
                f.write("Toplam Fiyat: ",toplamFiyat)
                f.write("\n")
                f.write("-"*50)
                cikis_kontrol = input("Baska bir islem yapmak istiyor musunuz? (evet-hayir) : ")
                if cikis_kontrol == "evet":
                    pass
                else:
                    print ("\n\nTEŞEKKÜRLER")
                    break

main()
```
