# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random , time
class Kumanda:
    def __init__(self,televizyon_durum = "kapalı",ses_seviye = 50,kanal = "FOX",kanal_liste=["FOX","Kanal D","Habertürk","NTV","Show TV","Star TV"],televizyon_bilgi=["Philips","E-1234","Almanya",time.ctime(time.time())]): 
        self.televizyon_durum = televizyon_durum
        self.ses_seviye = ses_seviye
        self.kanal = kanal
        self.kanal_liste = kanal_liste
        self.televizyon_bilgisi = televizyon_bilgi
    def tv_acma(self):
        if (self.televizyon_durum == "kapalı"):
            print("Televizyon açılıyor....")
            self.televizyon_durum = "açık"
        else:
            print("Televizyon kapanıyor...")
            self.televizyon_durum = "kapalı"
            print("Televizyon açık")
    def tv_kapama(self):
        if self.televizyon_durum == "açık":
            print("Televizyon Kapanıyor...")
            self.televizyon_durum = "kapalı"
        else:
            print("Televizyon Zaten Kapalı")
    def ses_ayar(self):
        if self.televizyon_durum == "kapalı":
            print("Önce Televizyonu Açınız")
        else:
            print("1-Ses Yükseltme\n2-Ses Azaltma\n3-Ses Kapama\n4-Ses Açma")
            islem = int(input("Yapmak İstediğiniz İşlem: "))
            if not islem in range(1,5):
                print("Hatalı Tuşlama Yaptınız")
            elif islem == 1:
                while True:
                    print(f"Şu anki Ses Seviyesi : {self.ses_seviye}")
                    artirma = float(input(f"Ne Kadar Artırmak İstiyorsunuz (Max {100 - self.ses_seviye})"))
                    if self.ses_seviye + artirma > 100:
                        print("Ses Seviyesi Maximum 100 Olmalı")
                    else:
                        self.ses_seviye += artirma
                        print(f"Güncel Ses Seviyesi: {self.ses_seviye}")
                        break
            elif islem == 2:
                while True:
                    print(f"Şu anki Ses Seviyesi : {self.ses_seviye}")
                    azaltma = float(input(f"Ne Kadar Azaltmak İstiyorsunuz (Max {self.ses_seviye})"))
                    if self.ses_seviye - azaltma < 0:
                        print("Ses Seviyesi Minimum 0 Olmalı")
                    else:
                        self.ses_seviye -= azaltma
                        print(f"Güncel Ses Seviyesi: {self.ses_seviye}")
                        break
            elif islem == 3:
                self.ses_seviye = 0
                print("Ses Kapatıldı")
            elif islem == 4:
                if self.ses_seviye != 0:
                    print("Ses Zaten Açık")
                else:
                    self.ses_seviye = 50
                    print("Ses Seviyesi Normale Getirildi")
    def kanal_ekle(self):
        if self.televizyon_durum == "kapalı":
            print("Önce Televizyonu Açınız")
        else:
            kanal = input("Eklemek İstediğiniz Kanal Adı: ")
            if kanal in self.kanal_liste:
                print("Kanal Zaten Kayıtlı")
            else:
                uydu = input("Kayıtlı Olduğu Uydu Bilgileri: ")
                frekans = int(input("Frekans Bilgileri: "))
                sembol = int(input("Sembol Oranı: "))
                yd = input("Yatay(H)/Dikey(V): ")
                if not yd in "HhVv":
                    print("Hatalı Giriş")
                self.kanal_liste.append(kanal)
                print("Kanal Eklendi")
                print(self.kanal_liste)
    def kanal_sayisi(self):
        if self.televizyon_durum == "kapalı":
            print("Önce Televizyonu Açınız")
        else:
            print(f"Bu Televizyonda {len(self.kanal_liste)} kanal bulunmaktadır")
    def kanal_gecme(self):
        l = list(self.kanal_liste)
        if self.televizyon_durum == "kapalı":
            print("Önce Televizyonu Açınız")
        else:
            rastgele = random.randint(0,len(l) - 1)
            print(f"{l[rastgele]} Kanalına Geçildi")

    def tv_bilgi(self):
        if self.televizyon_durum == "kapalı":
            print("Önce Televizyonu Açınız")
        else:
            l = self.televizyon_bilgisi
            print(f"Marka: {l[0]}\nModel: {l[1]}\nÜretim Yeri: {l[2]}\nTarih: {l[3]}")

kumanda = Kumanda()

print("İşlem Menüsü".center(50,"-"))
print("1-Tv Açma\n2-Tv Kapama\n3-Ses Ayarları\n4-Kanal Ekleme\n5-Kanal Sayısı Öğrenme\n6-Rastgele Kanala Geçme\n7-Televizyon Bilgileri\nÇıkış: 'q'")
while True:
    islem = input("Yapacağınız İşlemi Seçiniz: ")
    if islem == "1":
        kumanda.tv_acma()
    elif islem == "2":
        kumanda.tv_kapama()
    elif islem == "3":
        kumanda.ses_ayar()
    elif islem == "4":
        kumanda.kanal_ekle()
    elif islem == "5":
        kumanda.kanal_sayisi()
    elif islem == "6":
        kumanda.kanal_gecme()
    elif islem == "7":
        kumanda.tv_bilgi()
    elif islem == "q" or islem == "Q":
        print("Çıkış Yapıldı")
        break
    else:
        print("Hatalı Tuşlama Yaptınız")
```
